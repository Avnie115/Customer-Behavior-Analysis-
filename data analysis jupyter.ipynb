we will find the median(because it is not affected by outliers as mean is ) to  replace null values in review rating , but we will not simply replace every null value with median of the column but , we will first find the median of each category and replace the missing values with median of the category it belongs to 
so that that if review rating for footwear category is missing then we can replace it by median of footwear review rating and for clothes category we use clothes review rating to calculate median and put in place of missing value for clothes

```python
 df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

now to keep the code consistent and be able to refer code easily without running into errors we will transforma all column names into snake case

Customer ID ---> customer_id everything in lower case and words connected with underscores

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

We will now do some FEATURE ENGINEERING creating new columns and deleting the one's not needed

We will create a column age group diving age into 4 categories young Adult , Adult , Midlle-aged , Senior

```python
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

we will create new column purchase_frequency_days since the frequency_of_purchases column is a textual column (quaterly , weekly , monthly etc) we will convert it to numerical for better analysis

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

discount_applied and promo_code_used seem to be similar columns let's see if any one is redundant and we can drop one or not

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

since both the columns have same value we can drop one , let's drop promo_code_used

```python
df = df.drop('promo_code_used',axis =1)
```
